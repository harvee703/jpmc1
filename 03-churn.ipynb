```python
import pandas as pd
import numpy as np

import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/fenago/MLEssentials2/main/datasets/WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

```python
len(df)
```

## Initial data preparation

```python
df.head()
```

```python
df.head().T
```

```python
df.dtypes
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df['TotalCharges'] = df['TotalCharges'].fillna(0)
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')

string_columns = list(df.dtypes[df.dtypes == 'object'].index)

for col in string_columns:
    df[col] = df[col].str.lower().str.replace(' ', '_')
```

```python
df.churn = (df.churn == 'yes').astype(int)
```

```python
df.head().T
```

```python
from sklearn.model_selection import train_test_split
```

```python
df_train_full, df_test = train_test_split(df, test_size=0.2, random_state=1)
```

```python
df_train, df_val = train_test_split(df_train_full, test_size=0.33, random_state=11)
```

```python
y_train = df_train.churn.values
y_val = df_val.churn.values
```

```python
del df_train['churn']
del df_val['churn']
```

## Exploratory data analysis

```python
df_train_full.isnull().sum()
```

```python
df_train_full.churn.value_counts()
```

```python
global_mean = df_train_full.churn.mean()
round(global_mean, 3)
```

```python
categorical = ['gender', 'seniorcitizen', 'partner', 'dependents',
               'phoneservice', 'multiplelines', 'internetservice',
               'onlinesecurity', 'onlinebackup', 'deviceprotection',
               'techsupport', 'streamingtv', 'streamingmovies',
               'contract', 'paperlessbilling', 'paymentmethod']
numerical = ['tenure', 'monthlycharges', 'totalcharges']
```

```python
df_train_full[categorical].nunique()
```

```python
df_group = df_train_full.groupby(by='gender').churn.agg(['mean'])
df_group['diff'] = df_group['mean'] - global_mean
df_group['risk'] = df_group['mean'] / global_mean
df_group
```

```python
from IPython.display import display
```

```python
global_mean = df_train_full.churn.mean()
global_mean
```

```python
for col in categorical:
    df_group = df_train_full.groupby(by=col).churn.agg(['mean'])
    df_group['diff'] = df_group['mean'] - global_mean
    df_group['risk'] = df_group['mean'] / global_mean
    display(df_group)
```

```python
from sklearn.metrics import mutual_info_score
```

```python
def calculate_mi(series):
    return mutual_info_score(series, df_train_full.churn)

df_mi = df_train_full[categorical].apply(calculate_mi)
df_mi = df_mi.sort_values(ascending=False).to_frame(name='MI')


display(df_mi.head())
display(df_mi.tail())
```

```python
df_train_full[numerical].corrwith(df_train_full.churn).to_frame('correlation')
```

```python
df_train_full.groupby(by='churn')[numerical].mean()
```

## One-hot encoding

```python
from sklearn.feature_extraction import DictVectorizer
```

```python
train_dict = df_train[categorical + numerical].to_dict(orient='records')
```

```python
train_dict[0]
```

```python
dv = DictVectorizer(sparse=False)
dv.fit(train_dict)
```

```python
X_train = dv.transform(train_dict)
```

```python
X_train.shape
```

```python
feature_names = dv.feature_names_
feature_names
```

## Training logistic regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(solver='liblinear', random_state=1)
model.fit(X_train, y_train)
```

```python
val_dict = df_val[categorical + numerical].to_dict(orient='records')
X_val = dv.transform(val_dict)
```

```python
model.predict_proba(X_val)
```

```python
y_pred = model.predict_proba(X_val)[:, 1]
```

```python
y_pred
```

```python
churn = y_pred > 0.5
```

```python
(y_val == churn).mean()
```

## Model interpretation

```python
model.intercept_[0]
```

```python
dict(zip(dv.feature_names_, model.coef_[0].round(3)))
```

```python
subset = ['contract', 'tenure', 'totalcharges']
train_dict_small = df_train[subset].to_dict(orient='records')
dv_small = DictVectorizer(sparse=False)
dv_small.fit(train_dict_small)

X_small_train = dv_small.transform(train_dict_small)

dv_small.feature_names_
```

```python
model_small = LogisticRegression(solver='liblinear', random_state=1)
model_small.fit(X_small_train, y_train)
```

```python
model_small.intercept_[0]
```

```python
dict(zip(dv_small.feature_names_, model_small.coef_[0].round(3)))
```

```python
val_dict_small = df_val[subset].to_dict(orient='records')
X_small_val = dv_small.transform(val_dict_small)
```

```python
y_pred_small = model_small.predict_proba(X_small_val)[:, 1]
```
